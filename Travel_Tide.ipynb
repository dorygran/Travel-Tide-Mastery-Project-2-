```python
# Task 1. Ranking all our indexes in descending order, keeping NaNs:
data['rank_average_bags_scaled'] = data['average_bags_scaled'].rank(na_option = 'keep', ascending=False)
data['rank_bargain_hunter_index'] = data['bargain_hunter_index'].rank(na_option = 'keep', ascending=False)
data['rank_hotel_hunter_index_scaled'] = data['hotel_hunter_index_scaled'].rank(na_option = 'keep', ascending=False)
data['rank_cancelation_rate_scaled'] = data['cancelation_rate_scaled'].rank(na_option = 'keep', ascending=False)
data['rank_combined_booking_scaled'] = data['combined_booking_scaled'].rank(na_option = 'keep', ascending=False)
data['rank_session_count_index_scaled'] = data['session_count_index_scaled'].rank(na_option = 'keep', ascending=False)
data.head(5)
```

```python
# Task 2. minimum rating per user
data['min_c_h'] = data[['rank_average_bags_scaled', 'rank_bargain_hunter_index','rank_hotel_hunter_index_scaled',
                   'rank_cancelation_rate_scaled',  'rank_combined_booking_scaled',   'rank_session_count_index_scaled']].min(axis=1)
data.head()
```

```python
#Task 3. create a function to determine the rank based on the minimum rank encountered

def perk_define(row):
    if row['rank_hotel_hunter_index_scaled'] == row['min_c_h']:
        perk = 'Free Hotel Meal'
    elif row['rank_average_bags_scaled'] == row['min_c_h']:
        perk = 'Free Checked Bag'
    elif row['rank_cancelation_rate_scaled'] == row['min_c_h']:
        perk = 'No Cancellation Fee'
    elif row['rank_bargain_hunter_index'] == row['min_c_h']:
        perk = 'Exclusive Discount'
    elif row['rank_combined_booking_scaled'] == row['min_c_h']:
        perk = '1 Night Free Hotel with Flight'
    elif row['rank_session_count_index_scaled'] == row['min_c_h']:
        perk = 'Active Searcher'
    else:
        'Not Defined'
    return perk

```

```python
data['most_popular_perk'] = data.apply(perk_define, axis=1)
data.head()
```
